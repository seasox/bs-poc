```python
%pip install matplotlib
import matplotlib.pyplot as plt
import csv

# Read the CSV data
offsets = []
timings = []

with open('log/measurements.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        offset = int(row[0])
        timing = int(row[1])
        offsets.append(offset)
        timings.append(timing)
        
```

```python
print(timings)
```

```python

# Plot the data
plt.figure(figsize=(50, 6))
plt.step(offsets[15000:17000], timings[15000:17000])
plt.title('Timing vs Offset')
plt.xlabel('Offset')
plt.ylabel('Timing')
plt.grid(True)

# Show the plot
plt.show()
```
